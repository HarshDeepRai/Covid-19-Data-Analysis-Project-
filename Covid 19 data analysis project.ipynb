```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df
```

```python
#Active Caes=Confirmed-death-cured
covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail(5)
```

```python
statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise
```

```python
statewise["Recover rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.head()
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#top 10 active case state
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
top_10_active_cases
fig=plt.figure(figsize=(22,21))
plt.title('Top 10 state with most active cases',size=25)
ax=sns.barplot(data=top_10_active_cases.iloc[:10], y="Active_cases",x="State/UnionTerritory",linewidth=2,edgecolor='red')
plt.xlabel('States')
plt.ylabel('Total_active_cases')
plt.show()
```

```python
#top 10 state with highest death 
top_10_death_states=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values('Deaths',ascending=False).reset_index()
plt.figure(figsize=(20,10))
plt.title('Top 10 states with highest death rate')
ax=sns.barplot(data=top_10_death_states[:10],x='State/UnionTerritory',y='Deaths',linewidth=20)
plt.xlabel('States')
plt.ylabel('total_death')
plt.show()
```

```python
#growth rate in top 5 afftected state
fig=plt.figure(figsize=(20,20))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Uttar Pradesh','Karnataka','Kerala','Chhattisgarh'])],x='Date',y='Active_cases',hue='State/UnionTerritory')
```
